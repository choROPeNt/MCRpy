```python
import os
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
import fnmatch
import nrrd
```

```python
path = "/Volumes/data/160_10-layer/predictions/00/model_06-lowres-2024-01-29/subvolumes"
```

```python
for file in fnmatch.filter(os.listdir(path),"*.nrrd"):
    data,_  = nrrd.read(os.path.join(path,file))

    filename = os.path.splitext(file)[0]

    np.save(os.path.join(path,filename + ".npy"),data[150:350,:,:])
    
```
